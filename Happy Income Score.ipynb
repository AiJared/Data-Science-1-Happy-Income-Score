```python
# importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Loading dataset using pandas
dataset = pd.read_csv('happyscore_income.csv')
```

```python
#outputing the contents of the dataset
dataset
```

```python
# sorting the dataset using the average income
dataset.sort_values("avg_income", inplace=True)
dataset
```

```python
# filtering the dataset values in terms of satisfaction based on average income
satisfied = dataset[dataset['avg_income'] > 15000]
satisfied
```

```python
# checking the first and last(smallest) and last(largest) average incomes
satisfied.iloc[0]
satisfied.iloc[-1]
# printing the means satisfied average income and the dataset's average income
satisfied_mean = np.mean(satisfied['avg_income'])
all_mean = np.mean(dataset['avg_income'])
print(all_mean, satisfied_mean)
```

```python
# extra code to save the figures as high-res PNGs for the book
from pathlib import Path

IMAGES_PATH = Path() / "images" / "end_to_end_project"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

```python
# labeling the axes of the graph
plt.xlabel("income")
plt.ylabel("satisfaction")

# plotting the graph of income against satisfaction
plt.scatter(satisfied['avg_income'], satisfied['avg_satisfaction'])

# labelling the least income/least satisfied and the highest income/fairly satisfied country
plt.text(satisfied.iloc[0]['avg_income'],
        satisfied.iloc[0]['avg_satisfaction'],
        satisfied.iloc[0]['country'])

plt.text(satisfied.iloc[-1]['avg_income'],
        satisfied.iloc[-1]['avg_satisfaction'],
        satisfied.iloc[-1]['country'])
save_fig("income_satisfaction_graph")
plt.show()
```

```python
# labelling all the points on the graph
plt.xlabel("income")
plt.ylabel("satisfaction")

plt.scatter(satisfied['avg_income'], satisfied['avg_satisfaction'])

for k,row in satisfied.iterrows():
    plt.text(row['avg_income'],
            row['avg_satisfaction'],
            row['country'])
save_fig("income_satisfaction_labels")
plt.show()
```

```python
# Eyeballing satisfaction data
plt.xlabel("income")
plt.ylabel("satisfaction")
satisfaction = dataset["avg_satisfaction"]
income = dataset["avg_income"]
ineq = dataset['income_inequality']

plt.scatter(income, satisfaction, s=ineq*10, alpha=0.25)
save_fig("eyeballing_income_satisfaction_data")
plt.show()
```

```python
# Use Kmeans to interprete the data
from sklearn.cluster import KMeans

satisfied_income = np.column_stack((income, satisfaction))
km_res1 = KMeans(n_clusters=3).fit(satisfied_income)
clusters = km_res1.cluster_centers_

plt.xlabel("income")
plt.ylabel("satisfaction")

plt.scatter(income, satisfaction, s=100, alpha=0.25)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)
save_fig("km_res_for_satisfaction")
plt.show()
```

```python
# filtering dataset values based on the richest countries
richest = dataset[dataset["avg_income"] > 15000]
richest
```

```python
# mean and standard deviation of the average income of the richest countries
richest_mean = np.mean(richest["avg_income"])
std_dev_2 = np.std(richest["avg_income"])
print("The mean is: ",richest_mean,"\nThe standard deviation is: ", std_dev_2)
```

```python
# Eyeballing happy score data
happy = dataset["happyScore"]
income = dataset["avg_income"]
ineq = dataset['income_inequality']
#plt.scatter(income, ineq)
plt.xlabel("income")
plt.ylabel("happy")
plt.scatter(income, happy, s=ineq*10, alpha=0.25)
save_fig("eyeballing_happy_score_data")
plt.show()
```

```python
# Using KMeans to interprete the data
income_happy = np.column_stack((income, happy))

km_res2 = KMeans(n_clusters=3).fit(income_happy)

clusters = km_res2.cluster_centers_

plt.xlabel("income")
plt.ylabel("happy")
plt.scatter(income, happy, s=100, alpha=0.25)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)
save_fig("km_res_happy_score")
plt.show()
```
